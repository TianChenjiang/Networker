```python
import pandas as pd
import numpy as np
import glob
```

```python
path = r'D:\Study\Project\citi_cup\Networker\model\Experiment\data\fund' # use your path
all_files = glob.glob(path + "/*.csv")
```

```python
all_files
```

```python
target_path = r'D:\Study\Project\citi_cup\Networker\model\Experiment\data\fund_clean'
for file in all_files:
    fund_df = pd.read_csv(file, 
                      index_col=0, 
                      parse_dates=['ann_date', 'end_date']) 
    code = fund_df.loc[0]['ts_code']
    fund_df = fund_df.drop(['net_profit'], axis=1)
    fund_df = fund_df.interpolate()
    fund_df = fund_df.fillna(method='ffill')
    fund_df = fund_df.fillna(method='bfill')
    fund_df = fund_df.fillna(0)
    fund_df.to_csv(target_path + '\{}.csv'.format(code), index=False)
```
