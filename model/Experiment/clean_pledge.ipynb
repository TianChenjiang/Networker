```python
import pandas as pd
```

```python
pledge = pd.read_csv('data/pledge_with_enddate_new.csv')
```

```python
pledge.head()
```

```python
pledge.columns
```

```python
pledge = pledge.drop(['股东名称', '质押方', 
                      '冻结事由', '冻结期限', 
                      '最新收盘价(元)', '解除日期', 
                      '公告日期', '质押日期', 'Unnamed: 0'], axis=1)
```

```python
pledge.head()
```

```python
pledge.columns
```

```python
pledge = pledge.rename(columns = {
    '股票代码': 'code',
    '质押股数(万股)': 'this_pledge_nb',
    '占所持股比例(%)': 'this_share_ratio',
    '占总股本比例(%)': 'this_stock_ratio',
    '状态': 'status',
    '预警线(估算)': 'alarm_line',
    '平仓线(估算)': 'close_position_line',
    '质押日收盘价(前复权)(元)': 'price_on_pledge',
    '质押股数(万股).1': 'total_pledge_nb',
    '占所持股比例(%).1': 'total_share_ratio',
    '占总股本比例(%).1': 'total_stock_ratio',
    '平仓日': 'close_position_date'
})
```

```python
pledge.head()
```

```python
pledge.info()
```

```python
pledge['total_pledge_nb'] = pledge['total_pledge_nb'].apply(lambda x: str(x).replace(',', ''))
```

```python
import numpy as np
for i in range(pledge.shape[0]):
    for j in range(pledge.shape[1]):
        if pledge.iloc[i, j] == '--':
            pledge.iloc[i, j] = np.nan
```

```python
na_columns = ['total_pledge_nb', 'total_share_ratio', 'total_stock_ratio', 'this_pledge_nb']
for col in na_columns:
    pledge[col] = pd.to_numeric(pledge[col], errors='coerce')
    pledge[col] = pledge[col].fillna(pledge[col].mean())
```

```python
pledge.head()
```

```python
pledge['status'] = pledge['status'].apply(lambda x: 1)
```

```python
pledge.head()
```

```python
pledge['close_position_date'] = pledge['close_position_date'].apply(lambda x: x[:4] + '-' + x[4:6] + '-' + x[6:])
```

```python
pledge = pledge.drop([809])
```

```python
pledge['close_position_date'] = pd.to_datetime(pledge['close_position_date'])
```

```python
pledge.head()
```

```python
pledge['is_st'] = pledge.apply(lambda x: 1 if 'ST' in x.code else 0, axis=1)
```

```python
pledge['code'] = pledge['code'].apply(lambda x: x[2:] if 'ST' in x else x)
```
